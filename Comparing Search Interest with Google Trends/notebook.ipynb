## Trends for Analyst Jobs




```python
import pandas as pd
trends = pd.read_csv('datasets/multiTimeline.csv', skiprows=2)
trends_senior = pd.read_csv('datasets/senior_multiTimeline.csv', skiprows=2)
trends_jobs = pd.read_csv('datasets/jobs_multiTimeline.csv', skiprows=2)
trends_all = pd.read_csv('datasets/all_multiTimeline.csv', skiprows=2)
trends_all.head()
```

## Okay, that's great Google, but you are not making this data easily analyzable for us. I see a few things. Let's do the column names first. And I like lowercase, short column names.

```python
trends.columns = ["month", "financial analyst", "business analyst", 'data scientist']
trends.head()
```

Now, let's have an overall look.

```python
trends.info()
trends_senior.info()
```

<p>Now let's convert our <code>month</code> column from type object to datetime to make our date data more accessible.</p>

```python
trends.month = pd.to_datetime(trends.month)

trends.info()
trends.head()
```

## 6. Set month as index
<p>And finally, let's set the <code>month</code> column as our index to wrap our data cleaning. Having <code>month</code> as index rather than the zero-based row numbers will allow us to write shorter lines of code to create plots, where <code>month</code> will represent our x-axis.</p>

```python
trends = trends.set_index(['month'])
trends.head()
```

## 7. It's plot time now.
<p>We cleaned our data, we only need one line of code (and just <em>thirteen</em> characters!) to remake the Google Trends chart, plus another line to make the plot show up in our notebook.</p>

```python
%matplotlib inline
trends.plot()
trends_senior.plot()
trends_jobs.plot()
```

## 8. Business Analysis is new passion now!
<p>Is it enIt is not personal anymore. It is strickly business! There is so much to make sense of here. </p>
<p>Let's have a 48 months windows and see the rising of data science.</p>

```python
trends.loc['2020-03-01':].rolling(window=6).mean().plot()
trends_senior.loc['2020-03-01':].rolling(window=6).mean().plot()
trends_jobs.loc['2020-03-01':].rolling(window=6).mean().plot()
trends_all.loc['2020-03-01':].rolling(window=6).mean().plot()
```

## Long- term trends
<p>We can use rolling to highlight long-term trends and Those spikes are a sudden acceleration of search interest in a topic, compared to usual search volume. Let's make the window twelve months a.k.a. two years.</p>

```python
trends_all.rolling(window=24).mean().plot()
```
